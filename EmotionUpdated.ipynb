```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#  Step 3: Load Image
# Replace 'test.jpg' with the path to your image
image_path = "D:\\work\\emotion-detection-master\\happyBoy.jpg"

# Read and convert image
img_bgr = cv2.imread(image_path)
if img_bgr is None:
    raise Exception("Image not found. Please check the path.")

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Show image
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Original Image")
plt.show()
```

```python
emotion_detector = FER()
emotion, score = emotion_detector.top_emotion(img_bgr)
```

```python
print(f"Detected emotion: {emotion} with confidence score: {score}")
```

```python
import cv2
import numpy as np
from fer import FER
import time

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

cap = cv2.VideoCapture(1)  # External webcam (if available)

if not cap.isOpened():
    cap = cv2.VideoCapture(0)  # Built-in webcam

if not cap.isOpened():
    raise IOError("Cannot open webcam")

# Print confirmation
print("Webcam successfully opened!")

emotion_detector = FER()

# Optional: Add a short delay for camera stabilization
import time
time.sleep(2)  # Give the camera some time to warm up

while True:
  ret, frame = cap.read()
  #result = DeepFace.analyze(frame, actions=['emotion'])
  emotion, score = emotion_detector.top_emotion(frame)

  gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
  faces = face_cascade.detectMultiScale(gray,1.1, 4)
  for (x, y,w,h) in faces:
    cv2.rectangle(frame, (x,y), (x+w, y+h), (0,255,0), 2)

  font = cv2.FONT_HERSHEY_COMPLEX
  cv2.putText(
    frame,
    emotion,
    (0,50),
    font,3,
    (0,0,255),
    2, cv2.LINE_4
  );

  cv2.imshow("Webcam Face Detection", frame)

  if cv2.waitKey(2) & 0xFF == ord('q'):
    break

cap.release()
cv2.distroyAllWindows()
q
```
